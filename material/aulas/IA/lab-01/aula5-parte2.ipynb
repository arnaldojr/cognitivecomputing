### Aplicando GMM e DBSCAN


- `Gaussian Mixture Models` é uma abordagem probabilística para o agrupamento, que tenta modelar cada cluster como uma distribuição gaussiana (normal). 

- `DBSCAN (Density-Based Spatial Clustering of Applications with Noise)` é um algoritmo de clustering baseado na densidade dos pontos. Ele é capaz de identificar clusters de formas arbitrárias e tratar pontos ruidosos (outliers) de maneira eficiente.

    - O DBSCAN possui dois principais parâmetros:

        - `eps`: A distância máxima entre dois pontos para que sejam considerados vizinhos.
        - `min_samples`: O número mínimo de pontos vizinhos para que um ponto seja considerado parte de um cluster.


```python
#Importe das libs
import pandas as pd

from sklearn.cluster import KMeans, DBSCAN
from sklearn.mixture import GaussianMixture

import matplotlib.pyplot as plt
```

## Carregando os dados

```python
url_db = 'https://raw.githubusercontent.com/reisanar/datasets/master/moons.csv'
moons = pd.read_csv(url_db)
```

```python
moons.head()
```

```python
plt.scatter(x=moons['X'], y=moons['Y'])
```

## Criação de modelos de Agrupamento

```python
# Modelos de agrupamento

random_state = 42

#k-means
km = KMeans(n_clusters=4, random_state=random_state) # k=4
#Gaussian Mixture
gm = GaussianMixture(n_components=4, random_state=random_state) #k=4
#DBScan
db = DBSCAN(eps=0.4)
```

```python
#aplica o algoritimo e armazena o cluster de cada dado

km_c = km.fit_predict(moons)
gm_c = gm.fit_predict(moons)
db_c = db.fit_predict(moons)
```

```python
fig = plt.figure(figsize = (12,8), dpi=80)
plt.subplot(2,2,1)
plt.scatter(x=moons['X'],y=moons['Y'])
plt.title('Original')
plt.subplot(2,2,2)
plt.scatter(x=moons['X'],y=moons['Y'], c=km_c)
plt.title('KMeans')
plt.subplot(2,2,3)
plt.scatter(x=moons['X'],y=moons['Y'], c=gm_c)
plt.title('GaussianMixture')
plt.subplot(2,2,4)
plt.scatter(x=moons['X'],y=moons['Y'], c=db_c)
plt.title('DBScan')
```
